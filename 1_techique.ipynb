```python
from langchain.chains import LLMChain ## when prompt temp is combine with llm
from langchain import PromptTemplate

generic_temp="""
Write a summary of the following speech:
speech:{speech}
Translate the precise summary to "{language}
"""
```

```python
prompt=PromptTemplate(
    input_variables=['speech','language'],
    template=generic_temp
)
prompt
```

```python
complete_prompt=prompt.format(speech=speech,language='French')
complete_prompt
```

```python
llm.get_num_tokens(complete_prompt)
```

```python
llm_chain=LLMChain(llm=llm,prompt=prompt)
summary=llm_chain.run({'speech':speech,'language':'Hindi'})
summary
```

# StuffDocumentchain Text Summarization

```python
from langchain_community.document_loaders import PyPDFLoader

loader=PyPDFLoader("apjspeech.pdf")
docs=loader.load()
docs
```

```python
template="""
write a concise and short summary of the following specch,
Speech:{text}
"""
prompt=PromptTemplate(
    input_variables=['text'],
    template=template
)
```

```python
from langchain.chains.summarize import load_summarize_chain
chain=load_summarize_chain(llm,chain_type='stuff',prompt=prompt,verbose=True)
output_summary=chain.run(docs)
output_summary
```

# Map reduce to summarize large documents

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
docs
```

```python
final_documents=RecursiveCharacterTextSplitter(chunk_size=2000,chunk_overlap=100).split_documents(docs)
final_documents
```

```python
chunk_prompt="""
please summarize the below speech:
speech:'{text}'
summary:
"""

map_prompt_template=PromptTemplate(
    input_variables=['text'],
    template=chunk_prompt
)
```

```python
second_prompt_in_map="""
Provide the final summary of the entrie speech with these important points.
add a motivation title,start the precise summary with an introduction and provide the summary in number 
points for the speech.
Speech:{text}
"""
final_prompt_temp_in_map=PromptTemplate(input_variables=['text'],template=second_prompt_in_map)
final_prompt_temp_in_map
```

```python
summary_chain=load_summarize_chain(
    llm=llm,
    chain_type='map_reduce',
    map_prompt=map_prompt_template,
    combine_prompt=final_prompt_temp_in_map,
    verbose=True 
)

output=summary_chain.run(final_documents)
output
```

# Refine Chain For Summarization

```python
chain=load_summarize_chain(
    llm=llm,
    chain_type='refine',
    verbose=True
)
output_summary=chain.run(final_documents)
output_summary
```
